```python
import matplotlib.pyplot as plt
import seaborn as sns

from pandas import DataFrame
```

```python
data={'names':['Alice','Bob','Charlie','David','Eve'],
      'ages':[25,30,35,40,45],
      'salaries':[50000,60000,70000,80000,90000],
      'genders':['F','M','M','M','F']}
df=DataFrame(data)
df
```

```python
#Matplotlib bar chart
plt.bar(df['names'],df['ages'])
plt.xlabel('Names')
plt.ylabel('Ages')
plt.title('Ages of Employees')
```

```python
#seaborn bar chart
sns.barplot(x='names',y='ages',data=df)
plt.xlabel('Names')
plt.ylabel('Ages')
plt.title('Ages of Employees')
```

```python
#line plot matplotlib
plt.plot(df['names'],df['salaries'])
plt.xlabel('Names')
plt.ylabel('Salaries')
plt.title('Salaries of Employees')
```

```python
#line plot seaborn
sns.lineplot(x='names',y='salaries',data=df)
plt.xlabel('Names')
plt.ylabel('Salaries')
plt.title('Salaries of Employees')
```

```python
#pie chart matplotlib
plt.pie(df['ages'],labels=df['names'])
plt.title('Ages of Employees')
```

```python
#seaborn pie chart
colors = sns.color_palette('pastel')[0:5]
plt.pie(df['ages'], labels=df['names'], colors=colors)
plt.title('Ages of Employees')
```

```python
#defining elements of a plot
# plot element add context so the plot effectively conveys meaning to its viewers
#two methods:
#1.functional method
#2.object-oriented method
```

```python
#object-oriented method
#1.Create a blank figure object.
#2.add axes to the figure.
#3.Generate the plot within the  figure object
#4. specfy plotting and layout parameters for the plot within your figure.
```

```python
%matplotlib inline
#is a Jupyter Notebook command that tells Python to show graphs inside the notebook itself, just below the code, instead of opening them in a separate window.
rcParams['figure.figsize']=5,4
```

```python
x= range(1,10)
y= [1,2,3,4,0,4,3,2,1]
fig=plt.figure()#blank figure object
ax=fig.add_axes([.1,.1,1,1])
#ax=fig.add_axes([x_start , y_start , width , height])
#this fig.add_axes() method tells us where this graph will be placed in the empty page

#Think of it like this 
#fig = blank paper
#add_axes() = draw a rectangle on the paper
#ax = that rectangle where the graph goes
ax.plot(x,y)
```

```python
x= range(1,10)
y= [1,2,3,4,0,4,3,2,1]
fig=plt.figure()#blank figure object
ax=fig.add_axes([.1,.1,1,1])
ax.set_xlim(1,9)
ax.set_ylim(0,5)
ax.set_xticks([0,1,2,3,4,5,6,9,10])
ax.set_yticks([0,1,2,3,4,5])
ax.plot(x,y)
```

```python
x= range(1,10)
y= [1,2,3,4,0,4,3,2,1]
fig=plt.figure()#blank figure object
ax=fig.add_axes([.1,.1,1,1])
ax.set_xlim(1,9)
ax.set_ylim(0,5)
ax.grid()
ax.plot(x,y)
```

```python
#Generating multiple plots in one figure with subplots
fig= plt.figure()
fig, (ax1, ax2)= plt.subplots(nrows=1, ncols=2)
ax1.plot(x)
ax2.plot(x,y)
```

```python
fig=plt.figure()
ax=fig.add_axes([.1,.1,1,1])
mpg.plot(ax=ax)
ax.set_xticks(range(32))
ax.set_xticklabels(cars.carnames, rotation=60, fontsize='medium')
ax.set_title('Miles Per Gallon (MPG) of Cars')
ax.set_xlabel('Car Names')
ax.set_ylabel('Miles Per Gallon (MPG)')
ax.set_ylim(0,45)
ax.annotate('Highest MPG', xy=(19, 33.9), xytext=(20, 35), arrowprops=dict(facecolor='black', shrink=0.05))
ax.legend(loc='best')
```
