```python
#parametric correlation analysis:
#a method for finding correlation between linearly related continuous numeric variables
#R=1-> strong positive relationship
#R=-1-> strong negative relationship
#R=0-> no relationship/ not linearly correlated

#use Pearson Correlation
#to uncover(linear) realtionship between variables
#Not to rule out possible(nonlinear) relationships between variables
```

```python
#Pearson correlation coefficient
address = 'E:\Python_Lectures\data\mtcars.csv'
cars = pd.read_csv(address)
cars.columns=['','mpg','cyl','disp','hp','drat','wt','qsec','vs','am','gear','carb']
```

```python
sns.pairplot(cars)
```

```python
x=cars[['mpg','hp','qsec','wt']]
sns.pairplot(x)
```

```python
#using scipy to calculate the pearson correlation coefficient
mpg=cars.mpg
hp=cars.hp
qsec=cars.qsec
wt=cars.wt
pearsonr_rcoefficient, p_value= pearsonr(qsec, wt)
print('pearson r coefficient: ', pearsonr_rcoefficient)
```

```python
#using pandas to calculate the pearson correlation coefficient
corr= x.corr()
corr
```

```python
#using seaborn to visualize the pearson correlation coefficient
sns.heatmap(corr, xticklabels=corr.columns.values, yticklabels=corr.columns)
```

```python
#Nonparametric correlation Analysis:
#you can use nonoarametric correlation analysis to find correlation between categorical, nonlinearly related, non-normally distibuted variables
#1,Spearman's rank correlation coefficient
#spearman's rank correlation  finds the r correlation between variable pairs of the ordinal data types. Variables  pairs are then able to be ranked according to the strength according to the strenght of the correlation between them.
#R=1-> strong postive relationship
#r=0-> no relationship
#R=-1-> strong negative relationship
#The spearman's correlation assumes:
#Your variable are ordinal(in other words, numeric but able to be ranked like a categorical variable)
#Your variable are related nonlinearly
#your data is not normally distributed


#Chi-square tables test for Independence:
#p<0.05-> reject null hypothesis and conclude that the variables are correlated
#p>0.05-> accespts null hypothesis and conclude that the variables are independent

#When to use chi-square Tables:
#your variables are categorical or numeric
#you have binned the numeric variables


```

```python
address = 'E:\Python_Lectures\data\mtcars.csv'
cars = pd.read_csv(address)
cars.columns=['car names','mpg','cyl','disp','hp','drat','wt','qsec','vs','am','gear','carb']
cars.head()
```

```python
sns.pairplot(cars)
```

```python
x=cars[['mpg','hp','qsec','wt']]
sns.pairplot(x)
```

```python
cyl=cars.cyl
vs=cars.vs
am=cars.am
gear=cars.gear

spearmanr_coefficient, p_value= spearmanr(cyl, gear)
print('spearmanr coefficient: ', spearmanr_coefficient)
```

```python
spearmanr_coefficient, p_value= spearmanr(vs, gear)
print('spearmanr coefficient: ', spearmanr_coefficient)
```

```python
#chi-square test for independence
table=pd.crosstab(gear, cyl)
from scipy.stats import chi2_contingency
chi2, p, dof, expected = chi2_contingency(table.values)
print('p-value: ', p)
print('chi2: ', chi2)
print('degrees of freedom: ', dof)
print('expected frequencies: ', expected)
```
