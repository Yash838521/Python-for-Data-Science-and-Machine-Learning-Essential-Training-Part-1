Cleaning and Treating Categorical Variables

```python
import numpy as np
from pandas import  DataFrame
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
data={'names': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Yash','kunal'],
      'age': [25, 30, 35, 40, 45, 50, 55],
      'gender': ['F', 'M', 'M', 'M', 'F', 'M', 'M'],
      'rank': [2, 1, 3, 4, 5, 6, 7]}
df= DataFrame(data)
df
```

```python
#Label Encoding
lable_encoder= LabelEncoder()
lable_encoder.fit(df['names'])
```

```python
label_encoded_names= lable_encoder.transform(df['names'])
print('Label Encoded Names: ', label_encoded_names)
```

```python
#one hot encoder;
onehot_encoder= OneHotEncoder(sparse_output=False)
```

```python
onehot_encoder.fit(df[['names']])
```

```python
import pandas as pd
onehot_encoded_names = onehot_encoder.transform(df[['names']])

# Convert to DataFrame
onehot_encoded_df = pd.DataFrame(
    onehot_encoded_names,
    columns=onehot_encoder.get_feature_names_out(['names'])
)

# Add original column (optional)
onehot_encoded_df['names'] = df['names']

onehot_encoded_df
```
