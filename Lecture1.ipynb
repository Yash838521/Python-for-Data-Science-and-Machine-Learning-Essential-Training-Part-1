```python
#missing values in python
#by default, missing values are represented with NaN,"not a number"
data={'names':['yash', 'micheal','josh','abrish', 'kunal', 'Amit'],
     'age':[20,22,23,25,26,24],
     'gender':['male','male','male','male','male','male'],
     'rank':[2,3,4,5,6,7]}
ranking_df= DataFrame(data)
ranking_df.iloc[2:5, 1]=np.nan
ranking_df.iloc[3:6, 3]=np.nan
ranking_df.iloc[3,:]=np.nan
ranking_df
```

```python
#handling missing values
ranking_df.isnull()
```

```python
ranking_df.notnull()
```

```python
bool_series=pd.isnull(ranking_df['age'])
ranking_df[bool_series]# it will return only those values who's ahe is null
```

```python
ranking_df.fillna(0)
```

```python
ranking_df.fillna(method='pad') # this will replace the null values with it's prior values
```

```python
ranking_df.fillna(method='bfill') # null values will be flled with the post values with it
```

```python
ranking_df.interpolate(method='linear') # lineraly increase the  numerical values
```

```python
ranking_df.dropna() # all which has null values there field are droped
```

```python
ranking_df.dropna(how='all') # this drops the null values which has all field null
```

```python
ranking_df.dropna(axis=1)# this will drop the values whos rows vlaues are missing
```

```python
ranking_df.dropna(axis=0)# this will drop the values whos cols vlaues are missing
```

```python
df_obj1= DataFrame(np.arange(15).reshape(5,3))
df_obj1
```
