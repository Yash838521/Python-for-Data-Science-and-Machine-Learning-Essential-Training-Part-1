```python
import difflib
```

```python
fline='Hello. How are you? I am fine'
glines= 'How are you, lillian? I am doing well'
```

```python
d=difflib.Differ()
diff= d.compare(fline, glines)
for line in diff:
    print(line)

#Here with this diiffer() function, we are spliting the  two lines based on there similirity and there differences. 
#The variables with having '-' sign are the ones which are present in the first line but not in the second line.
#The variables with having '+' sign are the ones which are present in the second line but not in the first line.
#The variables with having no sign are the ones which are present in both the lines.
```

Four Beautiful Soup Object Types
1.BeautifulSoup object
2.Tag object
3.NavigableString Object
4.Comment object


BeautifullSoup object:
1.The BeautifulSoup object is a representation of the document you're scraping as a whole.
2.It is easily navigable and Searchable

The Tag Object:
1.Tag object correspond to XML and HTML elements in an original document
2.You can navigate and refference dat tag attributes

The NavigableString:
1.A NavigableString object is used to add a bit of text within a tag.
2.Beautiful soup uses the NavigableString class as a container for bits of text

The Comment object:
The comment object is a type of NavigableString that you can use for commenting your code

```python
#Asynchronous web scraping
import asyncio
import aiohttp
from bs4 import BeautifulSoup
import re
import csv
```

```python
!pip install nest-asyncio
import nest_asyncio
nest_asyncio.apply()
```

```python
async def scrap_and_save_links(text):
    soup= BeautifulSoup(text, 'html.parser')
    file=open('csv_file', 'a', newline='')
    writer= csv.writer(file, delimiter=',')
    for link in soup.find_all('a', attrs={'href': re.compile('^https')}):
        soup_link= str(link)
        writer.writerow([soup_link])
    file.close()
        
```

```python
async def fetch(session, url):
    try:
        async with session.get(url) as response:
            text= await response.text()
            task= asyncio.create_task(scrap_and_save_links(text))
            await task
    except Exception as e:
                print(str(e))
                      
```

```python
async def scrap(urls):
    task=[]
    async with aiohttp.ClientSession() as session:
        for url in urls:
            task.append(fetch(session, url))
        await asyncio.gather(*task)
        
```

```python
urls=['https://analytics.usa.gov', 'https://www.python.org', 'https://www.wikipedia.org']
asyncio.run(scrap(urls=urls))
```
